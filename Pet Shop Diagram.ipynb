We designed our table to look specifically at each individual transaction.  Each pet and fish have their own ID and each pet, fish, and food have their own order form.  The tables include information about the animals’ species, names, and colors as well as the ID of their aquarium or enclosure.  This allows the pet shop to keep better track of the exact pet or fish sold because they can have a database of all the current pets and fish and all the ones that have been sold.  The animals can all be connected to their specific enclosures and aquariums so that the pet shop can see which enclosures are full or empty at any time.  The databases on the enclosures and aquariums include information on their location in the pet shop.  Included is also an expense report for food and an order database for food.  The food expense report includes the type of food, the cost to buy it, the cost to sell it, the amount that has been bought, and the amount that has been sold.  All these orders are tied to the customer’s ID that includes information about their names and their loyalty program ID if they have one.  There is a separate table as well that lists the customer ID of the different loyalty program IDs.  The central point of the diagram is the customer table since it has the most other tables connected to it and all the other tables can be traced back from it.
